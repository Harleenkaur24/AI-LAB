```python
import numpy as np

random_array = np.random.randint(0, 1001, size=365)

print(random_array)
```

```python
w= int(input("Enter the window size (w):"))
s= int(input("Enter the stride (s):"))
b= float(input("Enter the beta value (b):"))
```

```python
windows = []
i = 0

while i < len(random_array):
    window = random_array[i:i + w]

    if len(window) < w:
        window = np.pad(window, (0, w - len(window)), mode='constant', constant_values=0)

    windows.append(window)
    i += s

print(f"Sliding Windows with window size {w} and stride {s}:")
for idx, window in enumerate(windows):
    print(f"Window {idx + 1}: {window}")
```

```python
mean_vector= []
max_vector= []
min_vector= []

for window in windows:
    mean=np.mean(window)
    max=np.max(window)
    min=np.min(window)

    mean_vector.append(mean)
    max_vector.append(max)
    min_vector.append(min)

print("Mean vector:", mean_vector);
print("Max vector:", max_vector);
print("Min vector:", min_vector);
```

```python
len(mean_vector)
```

```python
moving_average_vector=[]

moving_average_vector.append(mean_vector[0])

for t in range(1,len(mean_vector)):
  current_value= mean_vector[t]
  previous_moving_average=moving_average_vector[-1]
  moving_average= b*(current_value) + (1-b)* previous_moving_average
  moving_average_vector.append(moving_average)

print("Moving Average:", moving_average_vector)
```

```python
mv_pos_cnt=0
mv_neg_cnt=0
mv_zero_cnt=0


for t in range(0, len(moving_average_vector)):
    current_value = moving_average_vector[t]
    if current_value > 0:
        mv_pos_cnt=mv_pos_cnt+1
    elif current_value <0:
        mv_neg_cnt=mv_neg_cnt+1

    else:
        mv_zero_cnt=mv_zero_cnt+1

print("Total no. of times moving average increases:", mv_pos_cnt)
print("Total no. of times moving average decreases:", mv_neg_cnt)
print("Total no. of times moving average remains zero", mv_zero_cnt)
```

```python
gradient_vector=[]

for t in range(0, len(moving_average_vector)-1):
  current_value=moving_average_vector[t]
  next_value=moving_average_vector[t+1]
  gradient=next_value - current_value
  gradient_vector.append(gradient)

print("Gradient:", gradient_vector)
```

```python
len(gradient_vector)
```

```python
pos_cnt=0
neg_cnt=0
zero_cnt=0


for t in range(0, len(gradient_vector)):
    current_value = gradient_vector[t]
    if current_value > 0:
        pos_cnt=pos_cnt+1
    elif current_value <0:
        neg_cnt=neg_cnt+1

    else:
        zero_cnt=zero_cnt+1

print("Total no. of times gradient increases:", pos_cnt)
print("Total no. of times gradient decreases:", neg_cnt)
print("Total no. of times gradient remains zero", zero_cnt)
```

```python
laplacian_vector=[]

for t in range(0, len(gradient_vector)-1):
  current_value=gradient_vector[t]
  next_value=gradient_vector[t+1]
  laplacian=next_value - current_value
  laplacian_vector.append(laplacian)

print("Laplacian:", laplacian_vector)
```

```python
lap_pos_cnt=0
lap_neg_cnt=0
lap_zero_cnt=0


for t in range(0, len(laplacian_vector)):
    current_value = laplacian_vector[t]
    if current_value > 0:
        lap_pos_cnt=lap_pos_cnt+1
    elif current_value <0:
        lap_neg_cnt=lap_neg_cnt+1

    else:
        lap_zero_cnt=lap_zero_cnt+1

print("Total no. of times laplacian increases:", lap_pos_cnt)
print("Total no. of times laplacian decreases:", lap_neg_cnt)
print("Total no. of times laplacian remains zero", lap_zero_cnt)
```

```python
def count_changes(data_vector, name=""):
    # Initialize counters
    pos_cnt = 0
    neg_cnt = 0
    zero_cnt = 0

    # Loop through the data vector
    for current_value in data_vector:
        if current_value > 0:
            pos_cnt += 1  # Increment positive count
        elif current_value < 0:
            neg_cnt += 1  # Increment negative count
        else:
            zero_cnt += 1  # Increment zero count

    # Print the results
    print(f"Total no. of times {name} increases:", pos_cnt)
    print(f"Total no. of times {name} decreases:", neg_cnt)
    print(f"Total no. of times {name} remains zero:", zero_cnt)


# User input for what to display
user_choice = input("Enter which vector to analyze ( 'gradient' or 'laplacian'): ").strip().lower()

if user_choice == "gradient":
    count_changes(gradient_vector, "gradient")
elif user_choice == "laplacian":
    count_changes(laplacian_vector, "laplacian")
else:
    print("Invalid input! Please enter 'gradient' or 'laplacian'.")
```
